```python
import json
import os
from concurrent.futures import ProcessPoolExecutor, as_completed

def process_document(doc, word_list):
    result = {}
    with open(doc, 'r') as book:
        book_info = json.load(book)
    for word in word_list:
        doc_freq_info = {
            'name': book_info['name'],
            'examples': set(),
            'count': 0
        }
        lines = book_info['lines']
        for shlok in lines:
            if word in shlok['root_words']:
                if len(doc_freq_info['examples']) < 3:
                    doc_freq_info['examples'].add(shlok['text'])
                doc_freq_info['count'] += 1
        if doc_freq_info['count'] > 0:
            doc_freq_info['examples'] = list(doc_freq_info['examples'])
            if word not in result:
                result[word] = []
            result[word].append(doc_freq_info)
    return result

# Load the original dictionary
with open('other_jsons/sanskrit_dict.json', 'r') as json_file:
    word_dict = json.load(json_file)

word_list = list(word_dict.keys())

# List all JSON files in the directory except 'sanskrit_dict.json'
doc_list = [f for f in os.listdir() if (f.endswith('.json') and f != 'sanskrit_dict.json')]

count_dict = {}

# Parallel processing of documents
with ProcessPoolExecutor() as executor:
    future_to_doc = {executor.submit(process_document, doc, word_list): doc for doc in doc_list}
    for future in as_completed(future_to_doc):
        result = future.result()
        for word, doc_infos in result.items():
            if word not in count_dict:
                count_dict[word] = []
            count_dict[word].extend(doc_infos)

# Save the new dictionary to a file
with open('other_jsons/word_freq_2.json', 'w') as json_file:
    json.dump(count_dict, json_file, indent=4)
```
