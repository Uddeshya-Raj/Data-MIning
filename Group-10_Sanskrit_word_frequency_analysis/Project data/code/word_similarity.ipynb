```python
import json
from sentence_transformers import SentenceTransformer
import torch
from scipy.spatial.distance import cosine

import numpy as np
```

```python
import tf_keras as keras
```

```python
with open('other_jsons/sanskrit_dict.json', 'r') as json_file:
    word_dict = json.load(json_file)
```

```python
print(len(word_dict))
```

```python
new_word_dict = dict()
sentences_list = []
for word, meaning in word_dict.items():
    meaning_list = [m.strip() for m in meaning.split(";")]
    sentences_list += meaning_list
    new_word_dict[f'{word}'] = [len(meaning_list), meaning_list]
```

```python
len(sentences_list)
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
embeddings = model.encode(sentences_list, batch_size=1024, show_progress_bar=True)
```

```python
import copy
embeddings = embeddings.tolist()
tembeddings = copy.deepcopy(embeddings)
```

```python
embedding_dict = dict()
for key,value in new_word_dict.items():
    count = value[0]
    embedding_dict[f"{key}"] = embeddings[:count]
    del embeddings[:count]
```

```python
embedding_arr = []
for key,value in new_word_dict.items():
    count = value[0]
    for i in range(count):
        embedding_arr.append([key,tembeddings[i]])
    del tembeddings[:count]
```
